```python
import requests
import numpy as np
import pandas as pd
import logging
import csv
from time import sleep
from random import randint
from bs4 import BeautifulSoup 
```

```python
url = 'https://www.zolo.ca/toronto-real-estate/houses-for-rent'

#looping through URLs for each page

def scrap_page(url):
    address = []
    price = []
    bedrooms = []
    bathrooms = []
    r = requests.get(url, headers = {'User-agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    #sleep(randint(8, 15))
    c = r.content
    soup = BeautifulSoup(c,"html.parser")
    data = soup.find_all("article",{"class":"card-listing xs-relative xs-full-height xs-flex xs-flex-column-reverse xs-flex-justify-end rounded xs-overflow-hidden new"})
    for i in data:
        addresses = i.find("div",{"class":"card-listing--location"}).text
        address.append(addresses)
        propprice = i.find("li", {"class": "price xs-block xs-mb05 xs-text-2 bold"}).text
        price.append(propprice)
        propbed = i.find_all("li", {"class": "xs-inline xs-mr05"})[0].text
        bedrooms.append(propbed)
        propbath = i.find_all("li", {"class": "xs-inline xs-mr05"})[1].text
        bathrooms.append(propbath)
    return address, price, bedrooms, bathrooms

```

```python
#looping through pages

list_of_scrapped_pages = []
result_list = []
for i in range (1,20):
    updated_url = "https://www.zolo.ca/toronto-real-estate/houses-for-rent/page-" + str(i)
    #print("starting " + updated_url)
    result = scrap_page(updated_url)
    list_of_scrapped_pages.append(result)
    #print("ended " + str(i))
```

```python
df = pd.DataFrame(list_of_scrapped_pages, columns = ['address', 'price', 'beds', 'bath'])
df
```
