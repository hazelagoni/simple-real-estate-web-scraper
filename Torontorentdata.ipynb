```python
import requests
import numpy as np
import pandas as pd
import logging
import csv
from time import sleep
from random import randint
from bs4 import BeautifulSoup


url = 'https://www.zolo.ca/toronto-real-estate/houses-for-rent'

#looping through URLs for each page

def scrap_page(url):
    r = requests.get(url, headers = {'User-agent':'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
    #sleep(randint(8, 15))
    c = r.content
    soup = BeautifulSoup(c,"html.parser")
    articles = soup.find_all("article",{"class":"card-listing xs-relative xs-full-height xs-flex xs-flex-column-reverse xs-flex-justify-end rounded xs-overflow-hidden new"})

    listings = []
    for article in articles:
        listing = []  # I could also have used a named tuple so it is clear that the first thing in the listing is an address and then price and so on

        # the property at
        # listing[0] is the address
        # listing[1] is the price
        # listing[2] is the bedroom_type
        # listing[3] is the bathroom_type

        # add properties to the listing. one listing per article
        address = article.find("div", {"class": "card-listing--location"}).text
        price = article.find("li", {"class": "price xs-block xs-mb05 xs-text-2 bold"}).text
        bedroom_type = article.find_all("li", {"class": "xs-inline xs-mr05"})[0].text
        bathroom_type = article.find_all("li", {"class": "xs-inline xs-mr05"})[1].text
        listing.append(address)
        listing.append(price)
        listing.append(bedroom_type)
        listing.append(bathroom_type)
        listings.append(listing)
    return listings

#looping through pages
from pprint import pprint
list_of_scrapped_pages = []
result_list = []
for i in range (1,20):
    updated_url = "https://www.zolo.ca/toronto-real-estate/houses-for-rent/page-" + str(i)
    # print("starting " + updated_url)
    result = scrap_page(updated_url)
    pprint(result)
    answers = list_of_scrapped_pages.extend(result)
    # print("ended " + str(i))

pprint(list_of_scrapped_pages)
```
